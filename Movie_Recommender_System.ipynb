```python
import numpy as np
import pandas as pd
```

###Data Preprocessing


Check if there is any missing data in the above dataset

```python
movies.isnull().sum()
```

```python
#there are 3 movies for which the overview is not known, since overview is an important column so we are dropping those three columns
movies.dropna(inplace = True)
```

Check if there is any dupicate data in the dataset

```python
movies.duplicated().sum()
```

Now we will make another dataset with the columns 'movie_id', 'title', 'tags' and for tags column we will add four columns 'overview', 'genres', 'keywords', 'cast' and 'crew'         
And the info in these columns are weird so we need to convert it into proper format

```python
movies.iloc[0].genres
```

```python
# To convert genres column into proper format
import ast
def convert(obj):
  list1 = []
  for i in ast.literal_eval(obj):
    list1.append(i['name'])
  return list1
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

Now we have change the format of 'genres' into the proper format and the same thing we are going to do with the keyword, cast and crew column. In cast column, we are keeping the top 3 actors only

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#Cast column before
movies['cast'][100]
```

```python
# To convert cast column
def convert3(obj):
  list1 = []
  count=0
  for i in ast.literal_eval(obj):
    if count != 3:
      list1.append(i['name'])
      count+=1
    else:
      break
  return list1
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
# Crew column before
movies['crew'][1]
```

```python
# To convert crew column
def director(obj):
  list1 = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      list1.append(i['name'])
  return list1
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head()
```

```python
# Now the overview column is left which is a string actually and we will make it into lists and concat those lists so that it become a large list and again convert it into string which become a big paragraoh

# Overview column before
movies['overview'][2]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

Now, we will be creating a column named 'tags' which will be concantenation of overview, genres, cast, crew, keywords

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

At last we're gonna find similarities between the tags of two movies to ultimately recommend that to the end user. For this we use vectorization approach by converting the text into the vectors and this process is called text vectorization. 'Bag of Words' is the technique that we'll use in text vectorization.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
from IPython.utils.text import string
def stem(text):
  list2=[]
  for i in text.split():
    list2.append(ps.stem(i))
  string = " ".join(list2)
  return string
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

So we've converted every movie into a vector. Now we have to calculate the distance(Cosine distance i.e angle between vectors) from one movie to the other movie. The more the distance, the less the similarity and vice versa

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity[0]
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight')
```

```python
recommend('Spider-Man 3')
```
